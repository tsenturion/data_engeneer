```python
from pyspark.sql import SparkSession
import matplotlib.pyplot as plt

spark = SparkSession.builder.master("local[*]").appName("SparkVisual").getOrCreate()

data = [("Анна", 28), ("Иван", 35), ("Мария", 22), ("Павел", 30)]
df = spark.createDataFrame(data, ["Имя", "Возраст"])

# Преобразуем в pandas для построения графика
pdf = df.toPandas()

# Строим столбчатую диаграмму
plt.bar(pdf["Имя"], pdf["Возраст"], color="skyblue")
plt.title("Возраст участников")
plt.xlabel("Имя")
plt.ylabel("Возраст")
plt.show()
```
